# 1. Signal Construction and Generation

To fulfill the requirements of the study, a high-resolution 1D time series is constructed. The resolution is a critical factor because the signal contains pulses with a standard deviation as narrow as $\sigma = 0.01$.

### Parameters:
* **Time Range**: 0 to 100 units.
* **Binning ($dt$)**: $0.001$, resulting in $10^5$ samples. This resolution ensures that the narrowest pulse ($\sigma = 0.01$) is sampled with approximately 10 points across its core, preventing aliasing or loss of signal shape.
* **Signal Components**: The signal is a superposition of four Gaussian pulses placed at distinct positions in time, utilizing the widths $\sigma = [0.01, 0.1, 1, 10]$ as specified.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Task 1: Construct 1D time series
dt = 0.001
t = np.arange(0, 100, dt)

# Task 2: Generate Gaussian pulses
# Formula: f(t) = A * exp(-(t - t0)^2 / (2 * sigma^2))
sigmas = [0.01, 0.1, 1, 10]
positions = [10, 25, 45, 75]
amplitudes = [1.0, 1.0, 1.0, 1.0]

signal = np.zeros_like(t)
for s, p, a in zip(sigmas, positions, amplitudes):
    signal += a * np.exp(-(t - p)**2 / (2 * s**2))

# Visualization of the raw signal
plt.figure(figsize=(12, 4))
plt.plot(t, signal, label='Raw Signal')
plt.title("Generated 1D Signal with Varying Gaussian Widths")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

# 2. Convolution with Top-Hat Kernels

The signal is smoothed using a top-hat (boxcar) kernel of varying radii ($R$). The kernel is mathematically defined as:
$$K(t) = \frac{1}{2R} \text{ for } |t| \le R, \text{ else } 0$$

### Implementation Logic:
* **Normalization**: The kernel is normalized so that its integral equals unity. This preserves the total flux (area) of the Gaussian pulses, allowing for a direct comparison of how "detectability" (peak amplitude) changes.
* **Smoothing Process**: The convolution acts as a sliding window average. As the radius $R$ increases, the kernel incorporates more of the surrounding baseline, smoothing out high-frequency variations.

```python
# Task 3: Define top-hat kernels and convolve
def top_hat_kernel(radius, dt):
    # Kernel spans from -radius to +radius
    size = int(2 * radius / dt)
    if size % 2 == 0: size += 1  # Keep kernel centered
    kernel = np.ones(size) / size
    return kernel

radii = [0.05, 0.5, 5.0]
plt.figure(figsize=(12, 6))
plt.plot(t, signal, color='black', alpha=0.3, label='Raw Signal')

for R in radii:
    kernel = top_hat_kernel(R, dt)
    # Use mode='same' to keep output size equal to signal size
    convolved = np.convolve(signal, kernel, mode='same')
    plt.plot(t, convolved, label=f'Convolved (R={R})')

plt.title("Effect of Boxcar Convolution on Gaussian Pulses")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.xlim(0, 100)
plt.show()
```

# 3. Interpretation: $R$ vs. $\sigma$

The detectability of a pulse depends on the ratio between the kernel radius $R$ and the Gaussian width $\sigma$. There are three distinct regimes observed:

* **Regime 1: $R \ll \sigma$ (Signal Preservation)**
  * When the kernel is much narrower than the pulse, the convolution has a negligible effect. 
  * The peak amplitude and Gaussian shape are largely preserved.

* **Regime 2: $R \approx \sigma$ (Matched Smoothing)**
  * The smoothing begins to broaden the pulse significantly. 
  * The peak amplitude starts to drop as the energy is redistributed. This is often the regime used for "matched filtering" to maximize signal-to-noise in experimental data.

* **Regime 3: $R \gg \sigma$ (Signal Dilution)**
  * When the kernel is much wider than the pulse, the Gaussian behaves like a point source (delta function).
  * The output shape becomes a boxcar (matching the kernel), and the peak amplitude scales down by $\approx 1/R$. Detectability is severely reduced as the signal is "washed out" into the background.

```python
# Task 5: Compute Fourier Transform
N = len(t)
yf = fft(signal)
tf = fftfreq(N, dt)[:N//2]

plt.figure(figsize=(12, 5))

# Power Spectrum of Signal
plt.subplot(1, 2, 1)
plt.plot(tf, 2.0/N * np.abs(yf[:N//2]))
plt.xscale('log')
plt.yscale('log')
plt.title("Signal Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

# Spectrum of one Kernel
R_test = 0.5
kernel_full = np.zeros_like(t)
k_vals = top_hat_kernel(R_test, dt)
mid = len(t) // 2
kernel_full[mid - len(k_vals)//2 : mid + len(k_vals)//2 + 1] = k_vals
kf = fft(np.fft.fftshift(kernel_full))

plt.subplot(1, 2, 2)
plt.plot(tf, np.abs(kf[:N//2]))
plt.xscale('log')
plt.title(f"Boxcar Kernel Spectrum (R={R_test})")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Response")

plt.tight_layout()
plt.show()
```

# 4. Fourier Transform Analysis

Convolution in the time domain corresponds to multiplication in the frequency domain.

### Spectral Components:
* **Gaussian Pulse**: The Fourier transform of a Gaussian is also a Gaussian. There is an inverse relationship: the narrowest pulse ($\sigma = 0.01$) has the broadest frequency spectrum, extending to very high frequencies.
* **Top-Hat Kernel**: The Fourier transform of a boxcar kernel is a **Sinc function**:
  $$\mathcal{K}(f) = \frac{\sin(2\pi f R)}{\pi f}$$
* **Low-Pass Filtering**: The Sinc function acts as a low-pass filter. The "width" of the filter is inversely proportional to $R$. 

### Interpretation:
As $R$ increases, the filter becomes narrower in the frequency domain, cutting off high-frequency components. This explains why the sharpest pulses disappear first; their high-frequency power is suppressed by the "nulls" of the Sinc function, leaving only the broad, low-frequency features of the signal intact.

# 4. Flare Detection Analysis with Poisson Noise and Background

In this section, we transition from a deterministic signal to a stochastic one. In astroparticle physics, we often deal with discrete counts (photons, neutrinos, or cosmic rays). Therefore, the signal is better modeled as a realization of a **Poisson process**.

We will now:
1.  Redo the signal generation with an added **Poisson noise** component.
2.  Introduce a **slowly varying background**, which mimics diffuse emission or instrumental drift.
3.  Evaluate how the top-hat convolution helps in recovering flares (Gaussian pulses) from a noisy baseline.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# 1. Setup Time Series (High resolution)
dt = 0.01
t = np.arange(0, 100, dt)

# 2. Generate Theoretical "True" Signal (4 Gaussian Flares)
sigmas = [0.05, 0.2, 1.5, 8.0]
positions = [10, 30, 55, 80]
amplitudes = [15, 10, 8, 5] # Higher amplitudes to make Poisson stats visible

true_signal = np.zeros_like(t)
for s, p, a in zip(sigmas, positions, amplitudes):
    true_signal += a * np.exp(-(t - p)**2 / (2 * s**2))

# 3. Add Slowly Varying Background (e.g., a low-frequency trend)
# This represents diffuse sky background or seasonal variations
background = 3 + 2 * np.sin(2 * np.pi * t / 150)
total_mean_signal = true_signal + background

# 4. Apply Poisson Noise
# Each bin is a random draw from a Poisson distribution with mean = signal + bkg
noisy_signal = np.random.poisson(total_mean_signal)

# Visualization
plt.figure(figsize=(14, 5))
plt.step(t, noisy_signal, where='mid', color='gray', alpha=0.5, label='Detected Counts (Noisy)')
plt.plot(t, total_mean_signal, color='red', lw=2, label='True Mean (Signal + Bkg)')
plt.title("Simulated Blazar Flare Data: Poisson Noise & Varying Background")
plt.xlabel("Time")
plt.ylabel("Counts")
plt.legend()
plt.show()
```

### **Interpretation of Data Components**

*   **Poisson Noise:** This represents the intrinsic statistical fluctuation of detecting a finite number of events. Since $N$ detected counts have an uncertainty of $\sqrt{N}$, narrow flares with low amplitudes can easily be "hidden" within these fluctuations.
*   **Varying Background:** In real blazar observations, the "baseline" is rarely zero. It might change due to the movement of the telescope, atmospheric conditions, or long-term diffuse emission. This complicates flare detection because a local "peak" might just be the background rising.
*   **Blazar Analogy:** Each Gaussian pulse represents a "flare" — a sudden increase in flux from the object. Detecting these flares accurately requires smoothing techniques to distinguish real physics from random noise.

```python
# Define Top-Hat Kernels
def get_top_hat(radius, dt):
    size = int(2 * radius / dt)
    if size % 2 == 0: size += 1
    return np.ones(size) / size

radii_to_test = [0.1, 1.0, 5.0]

plt.figure(figsize=(14, 7))
plt.step(t, noisy_signal, where='mid', color='lightgrey', label='Raw Counts')

for R in radii_to_test:
    kernel = get_top_hat(R, dt)
    # Using 'same' mode to keep the time axis aligned
    convolved = np.convolve(noisy_signal, kernel, mode='same')
    plt.plot(t, convolved, label=f'Convolved (R={R})', lw=2)

plt.title("Smoothing Noisy Flare Data with Top-Hat Kernels")
plt.xlabel("Time")
plt.ylabel("Smoothed Counts")
plt.legend()
plt.show()
```

### **Detectability Analysis: Radius vs. Signal-to-Noise Ratio (SNR)**

When noise is present, the choice of kernel radius $R$ becomes a balancing act between **Signal-to-Noise Ratio (SNR)** and **temporal resolution**:

1.  **Small Radius ($R \approx 0.1$):** Effectively reduces bin-to-bin noise for narrow flares but leaves large fluctuations intact. The narrowest flare ($\sigma=0.05$) is visible, but the baseline remains very noisy.
2.  **Matched Radius ($R \approx \sigma$):** This is the optimal "Matched Filter" regime. By matching the kernel size to the expected flare width, we maximize the SNR. The noise is averaged out, while the flare amplitude is preserved better than with a larger kernel.
3.  **Large Radius ($R = 5.0$):** The background is beautifully smoothed, but the narrow flares are completely destroyed. Only the very broad flare at $t=80$ remains detectable, though its shape is now distorted into a flat-topped structure.

**Conclusion:** For a multi-scale signal like a blazar flare series, a single kernel radius is insufficient. Multi-scale smoothing or adaptive kernels are typically required to detect both "sharp" and "broad" flares simultaneously.

```python
# Fourier Transform Calculation
N = len(t)
yf_noisy = fft(noisy_signal)
tf = fftfreq(N, dt)[:N//2]

# Compute FFT of a sample kernel for comparison
R_sample = 1.0
kernel_full = np.zeros_like(t)
k_vals = get_top_hat(R_sample, dt)
mid = len(t) // 2
kernel_full[mid - len(k_vals)//2 : mid + len(k_vals)//2 + 1] = k_vals
kf = fft(np.fft.fftshift(kernel_full))

plt.figure(figsize=(12, 5))

# Plotting the Power Spectrum
plt.plot(tf, 2.0/N * np.abs(yf_noisy[:N//2]), label='Noisy Signal Spectrum', alpha=0.7)
plt.plot(tf, np.abs(kf[:N//2]) * 5, label='Kernel Response (Scaled)', color='red', lw=2)

plt.xscale('log')
plt.yscale('log')
plt.title("Frequency Domain Analysis of Noisy Flare Data")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Power")
plt.grid(which='both', linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

### **Spectral Interpretation**

*   **Poisson Noise (White Noise):** In the frequency domain, Poisson noise appears as a "floor" that spans all frequencies. This is visible as the flattened high-frequency power in the spectrum. 
*   **Slowly Varying Background:** The background trend contributes primarily to the very low-frequency power (the far left of the log-plot).
*   **Flare Signatures:** The Gaussian flares appear as peaks above the noise floor at specific frequency scales.
*   **The Filter Effect:** The Sinc function of the top-hat kernel (red line) acts as a low-pass filter. It suppresses the high-frequency noise floor while allowing the low-frequency signal to pass. However, any signal power located at frequencies higher than the first "null" of the kernel is lost, explaining why narrow flares disappear when $R$ is too large.

* **Background:** A Poissonian background level to simulate diffuse emission or instrumental noise.

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import SkyCoord
from gammapy.maps import MapAxis, WcsGeom, Map
from gammapy.datasets import MapDataset
from gammapy.modeling.models import (
    SkyModel,
    GaussianSpatialModel,
    PointSpatialModel,
    PowerLawSpectralModel,
    Models,
    FoVBackgroundModel,
)
from gammapy.estimators import ExcessMapEstimator, TSMapEstimator

# 1. Define the Geometry
energy_axis = MapAxis.from_energy_bounds("0.1 TeV", "10 TeV", nbin=1)
geom = WcsGeom.create(
    skydir=(0, 0), binsz=0.02, width=(5, 4), frame="galactic", axes=[energy_axis]
)

# 2. Define the Source Models
# Point Source
m1 = SkyModel(
    spatial_model=PointSpatialModel(lon_0="0.5 deg", lat_0="0.5 deg"),
    spectral_model=PowerLawSpectralModel(amplitude="1e-11 cm-2 s-1 TeV-1"),
    name="point_source",
)

# Extended Gaussian Sources
m2 = SkyModel(
    spatial_model=GaussianSpatialModel(lon_0="-0.5 deg", lat_0="-0.5 deg", sigma="0.1 deg"),
    spectral_model=PowerLawSpectralModel(amplitude="1e-11 cm-2 s-1 TeV-1"),
    name="ext_source_small",
)

m3 = SkyModel(
    spatial_model=GaussianSpatialModel(lon_0="1.0 deg", lat_0="-1.0 deg", sigma="0.3 deg"),
    spectral_model=PowerLawSpectralModel(amplitude="1e-11 cm-2 s-1 TeV-1"),
    name="ext_source_large",
)

models = Models([m1, m2, m3])

# 3. Create the Dataset and Simulate
# We assume a flat background for simplicity
dataset = MapDataset.create(geom=geom)
bkg_map = Map.from_geom(geom, data=10.0) # 10 counts per pixel background
dataset.background = bkg_map

# Add the source models to the dataset and simulate Poisson counts
dataset.models = models
dataset.fake()

# Plot the simulated counts
dataset.counts.sum_over_axes().plot(stretch="sqrt", add_cbar=True)
plt.title("Simulated 2D Counts Map")
plt.show()
```
